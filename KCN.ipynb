# Importing the Data

```python
from google.colab import files
import pandas as pd
uploaded=files.upload()
df=pd.read_csv('project3data.csv')
```

```python
# Since each article will have a minimum of one keyword , I decided to check nulls in the Keyword1 column which will clear all spaces between articles.
df = df[df['Keyword 1'].notna()]
```

```python
df.columns
```

```python
# Converting everything to lower case.
for column in df.columns:
  df[str(column)]=df[str(column)].str.lower()
```

```python
df
```

```python
# Creating a set to store every unique keyword.
sample=set()
for i in range(1,13):
  x=set(df.iloc[:,i].unique())
  sample=sample.union(x)
```

```python
sample
```

```python
# Removing null value.
import numpy as np
sample.remove(np.nan)
```

```python
len(sample)
```

```python
# A zero matrix 
A=np.zeros((len(sample),len(sample)),dtype=int)
```

```python
A
```

```python
sample=list(sample)
```

```python
sample.sort()
sample
```

```python
type(A)
```

```python
# Converting matrix to a dataframe.
f=pd.DataFrame(A)
```

```python
# Assign unique keywords as column and row index.
f.columns=sample
f.index=sample
```

```python
f
```

# Python code to create adjacency matrix.

```python
# A for loop to create the adjacency matrix.
for x in f.columns:
  for y in f.index:
    count=0
    for z in range(0,len(df)):
      if x==y:
        f[str(y)][str(x)]=0
      elif ((x in list(df.iloc[int(z),1:])) & (y in list(df.iloc[int(z),1:]))):
        count+=1
    f[str(y)][str(x)]=count
f
```

```python
f.index
```

```python
f.columns
```

```python
# Converting dataframe to numpy matrix.
mat=f.to_numpy()
```

```python
mat.shape
```

# Creating Weighted Graph

```python
import networkx as nx
G=nx.from_numpy_matrix(mat,parallel_edges=False)
pos=nx.spring_layout(G)
nx.draw(G,pos)
labels=nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
```

# Node Weight

```python
 G.degree(weight='weight')
```

# Node Degree

```python
 G.degree
```

```python
degree=[]
for k,v in G.degree:
  degree.append(v)
```

```python
strength=[]
for k,v in G.degree(weight='weight'):
  strength.append(v)
```

```python
df_final=pd.DataFrame(columns=['Keyword','degree','strength'])
```

```python
len(strength)
```

```python
df_final['Keyword']=sample
df_final['degree']=degree
df_final['strength']=strength
```

```python
# Final dataframe with keyword , degree and its strength
df_final
```

```python
# Taking top 30 values so I can remove duplicates and attenuate the data to accomodate only top 10 values.
largest=dict(f.max().nlargest(30))
```

```python
pair=pd.DataFrame(columns=['Keyword1','Keyword2','weight'])
```

```python
key1=[]
nodeweight=[]
for k,v in largest.items():
  key1.append(k)
  nodeweight.append(v)
```

```python
key2=[]
for k,v in largest.items():
  key2.append(f[f[k].where(f[k]==v).notnull()].index[0])
```

```python
pair['Keyword1']=key1
pair['Keyword2']=key2
pair['weight']=nodeweight
```
